```python
# Python Programming 

# List and function
# Finish a function to user enter quarterlty sales amount.
# Finish a function to display sales
# In main fuction, invoke above functions and shw total and average sales
# Show the result as the example

QUARTER =["Spring", "Summer", "Fall", "Winter"]
           
def main():
    # create a list to hold the sales of four quarters 
    sales =[0.0] * len(QUARTER)

    inputSales(sales)
    
    print('\nDisplay the quartely sales')
    display (sales)
    
    print('\nDisplay the summary')
    print('Total sales:              $', sum(sales))
    print('Average quarterly sales   $', sum(sales)/len(QUARTER))

# Ask user to enter sales:
def inputSales(sales):
    for i in range(0, len(sales)):
        print('Enter the sales of', QUARTER[i] ,  ' : $' , end = ' ')
        sales[i] = float(input())

# 1) Finish the method here:
def display (sales):
    for i in range(len(sales)):
        print('The sales of',QUARTER[i],': $',sales[i])
 


        
# Invoke 
main()
```

```python
# 15) Create a bar chart to compare average Salary by gender

res = gameSalary.groupby("Gender")["Salary"].mean()

print(res)

res.plot(kind='bar', title="The Average Salary Female vs Male", color='blue')
```

```python
# 16+17) Create a bar chart to compare average Salary by Department

res = gameSalary.groupby("Department")["Salary"].mean()

print(res)

res.plot(kind='bar', title="Average Salary by Department", color='blue')
```

```python
# 17+18) Show number of employees for each performance level and create pie chart

res = gameSalary["Performance"].value_counts()

print(res)

res.plot(kind='pie', autopct='%1.1f%%', title="Employee Performance Distribution")
```

```python
# 19+20) Show number of employees for each department,
# sorted from high to low and plot as bar chart

res = gameSalary["Department"].value_counts().sort_values(ascending=False)

print(res)

res.plot(kind='bar', title="Number of Employees by Department", color='blue')
```
